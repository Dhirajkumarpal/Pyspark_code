```
#project student data 

rdd24=sc.textFile('/FileStore/tables/StudentData.csv')
header = rdd24.first() #extract header
rdd24 = rdd24.filter(lambda row : row != header)  
```

```
rdd24.collect()
```

```
rdd25=rdd24.map(lambda x:x.split(',')[2])
rdd25.collect()
```

```
count_of_students=rdd25.distinct().count()
```

```
#total marks achieved by mail and femail
rdd26=rdd24.map(lambda x:(x.split(',')[1],x.split(',')[5]))
rdd26.collect()
```

```
rdd27=rdd26.reduceByKey(lambda x,y:int(x)+int(y))
```

```
rdd27.collect()
```

```
#number of students passed and failed
rdd28=rdd24.map(lambda x:(x.split(',')[2],x.split(',')[5]))
rdd28.collect()
```

```
rdd29=rdd28.filter(lambda x:int(x[1])>50)
rdd29.collect()
```

```
countofstudentspassed=rdd29.distinct().count()
```

```
#number of student per course
rdd30=rdd24.map(lambda x:(x.split(',')[3],x.split(',')[2]))
rdd30.collect()
```

```
rdd31=rdd30.distinct()
```

```
rdd32=rdd31.map(lambda x:(x[0],1))
```

```
rdd32.collect()
```

```
rdd33=rdd32.reduceByKey(lambda x,y:x+y)
```

```
rdd33.collect()
```

```
#marks of student per course
rdd34=rdd24.map(lambda x:(x.split(',')[3],x.split(',')[5]))
rdd34.collect()
```

```
rdd35=rdd34.reduceByKey(lambda x,y:int(x)+int(y))
rdd35.collect()
```

```
#average marks of students per course
rdd35=rdd24.map(lambda x:(x.split(',')[3],(x.split(',')[5],1)))
rdd35.collect()
```

```
rdd36=rdd35.reduceByKey(lambda x,y:(int(x[0])+int(y[0]),int(x[1])+int(y[1])))
rdd36.collect()
```

```
rdd37_average=rdd36.map(lambda x:(x[0],x[1][0]/x[1][1]))
rdd37_average.collect()
```
